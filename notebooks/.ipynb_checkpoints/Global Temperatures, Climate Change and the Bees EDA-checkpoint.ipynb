```python
# Get USA specific temperature data

USA_temp = country_temp[country_temp['Country'].isin(['United States'])]
print(USA_temp.columns)
USA_temp.describe()
```

```python
# Is the USA Warming? 
# code snippet adapted from https://www.kaggle.com/amelinvladislav/map-of-temperatures-and-analysis-of-global-warming
years = np.unique(USA_temp['dt'].apply(lambda x: x[:4]))
mean_temp_USA = []
mean_temp_USA_uncertainty = []

for year in years:
    mean_temp_USA.append(USA_temp[USA_temp['dt'].apply(
        lambda x: x[:4]) == year]['AverageTemperature'].mean())
    mean_temp_USA_uncertainty.append(USA_temp[USA_temp['dt'].apply(
        lambda x: x[:4]) == year]['AverageTemperatureUncertainty'].mean())
    
trace0 = go.Scatter(
    x = years,
    y = np.array(mean_temp_USA) + np.array(mean_temp_USA_uncertainty),
    fill = None,
    mode = 'lines',
    name = 'Uncertainty Upper Boundary',
    line = dict(
        color='rgb(255,0,255)',
    ),
)

trace1 = go.Scatter(
    x = years,
    y = np.array(mean_temp_USA) - np.array(mean_temp_USA_uncertainty),
    fill = 'tonexty',
    mode = 'lines',
    name = 'Uncertainty Lower Boundary',
    line = dict(
        color='rgb(255,0,255)',
    ),
)

trace2 = go.Scatter(
    x = years,
    y = mean_temp_USA,
    name = 'Average Temperature in the USA',
    line = dict(
        color = 'rgb(244, 241, 66)'
    )
)

data = [trace0,trace1,trace2]

layout = go.Layout(
    xaxis=dict(title='year'),
    yaxis=dict(title='Average Temperature, Celcius'),
    title='Average Temperature in the United States',
    showlegend=False)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

# Quick Analysis:

From the chart above we can see that the average uncertainty about the temperature is almost unable to be visualized in the United States as we approacht he start of the 21st century. This is the complete opposite of my expectations Since the Average temperature has been steadily growing for the last century, and because of the mass hysteria surrounding the growing concern for our Climate.
